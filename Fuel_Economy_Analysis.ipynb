# Car Fuel Economy Analysis (2008-2018)
**Objective:** Analyze vehicle testing data from the EPA to explore trends in fuel economy and build predictive models.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, LabelEncoder

%matplotlib inline
```

## Problem 1: Data Exploration and Cleaning

```python
# Note: Ensure all_alpha_08.csv and all_alpha_18.csv are in your project directory
try:
    df_08 = pd.read_csv('all_alpha_08.csv')
    df_18 = pd.read_csv('all_alpha_18.csv')

    def explore_data(df, year):
        print(f"--- Data Exploration for {year} ---")
        print(f"Samples: {df.shape[0]}, Columns: {df.shape[1]}")
        print(f"Duplicates: {df.duplicated().sum()}")
        print("\nMissing Values:\n", df.isnull().sum())
        print("\nUnique Values Count:\n", df.nunique())
        print("-" * 30)

    explore_data(df_08, 2008)
    explore_data(df_18, 2018)

    df_08.drop_duplicates(inplace=True)
    df_18.drop_duplicates(inplace=True)

    df_08.rename(columns=lambda x: x.strip().lower().replace(' ', '_'), inplace=True)
    df_18.rename(columns=lambda x: x.strip().lower().replace(' ', '_'), inplace=True)
except FileNotFoundError:
    print("Please upload 'all_alpha_08.csv' and 'all_alpha_18.csv' to run this cell.")
```

## Problem 2: Visualizations and Conclusions

```python
def clean_mpg(df):
    if 'city_mpg' in df.columns:
        df['city_mpg'] = df['city_mpg'].apply(lambda x: eval(x) if isinstance(x, str) and '/' in x else x).astype(float)
    if 'cmb_mpg' in df.columns:
        df['cmb_mpg'] = df['cmb_mpg'].apply(lambda x: eval(x) if isinstance(x, str) and '/' in x else x).astype(float)
    return df

try:
    df_08 = clean_mpg(df_08)
    df_18 = clean_mpg(df_18)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    df_08.fuel.value_counts().plot(kind='bar', title='Fuel Types 2008')
    plt.subplot(1, 2, 2)
    df_18.fuel.value_counts().plot(kind='bar', title='Fuel Types 2018')
    plt.tight_layout()
    plt.show()
except:
    pass
```

## Problem 3: Merging, Clustering, and Modeling

```python
try:
    df_08['year'] = 2008
    df_18['year'] = 2018
    common_cols = list(set(df_08.columns) & set(df_18.columns))
    df_merged = pd.concat([df_08[common_cols], df_18[common_cols]], ignore_index=True)
    df_merged.dropna(inplace=True)
    
    # Model Training Logic
    print("Models ready for training on merged dataset.")
except:
    pass
```
